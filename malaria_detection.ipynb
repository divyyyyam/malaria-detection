```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shahriar26s/malaria-detection")

print("Path to dataset files:", path)
```

```python
import os, shutil, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG19
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2

# -----------------------------
# ✅ Dataset Paths
# -----------------------------
# Update paths to use the downloaded dataset location
train_dir = os.path.join(path, "Malaria Dataset", "train")
val_dir   = os.path.join(path, "Malaria Dataset", "valid")
test_dir  = os.path.join(path, "Malaria Dataset", "test")

print("✅ Train dir :", train_dir)
print("✅ Val dir   :", val_dir)
print("✅ Test dir  :", test_dir)

# -----------------------------
# ✅ Data Generators
# -----------------------------
BATCH_SIZE = 32
IMG_SIZE = (224, 224)

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='binary'
)
val_generator = val_test_datagen.flow_from_directory(
    val_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='binary'
)
test_generator = val_test_datagen.flow_from_directory(
    test_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='binary', shuffle=False
)

# -----------------------------
# ✅ Build VGG19 Model
# -----------------------------
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = False  # First stage: feature extraction

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid', kernel_regularizer=l2(0.001))(x)

model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# -----------------------------
# ✅ Early Stopping
# -----------------------------
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# -----------------------------
# ✅ Training Stage 1 (Feature Extraction)
# -----------------------------
history = model.fit(train_generator, validation_data=val_generator, epochs=5, callbacks=[early_stop])

# -----------------------------
# ✅ Fine-Tuning Stage 2
# -----------------------------
for layer in base_model.layers[-4:]:  # Unfreeze last 4 layers
    layer.trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])

history_fine = model.fit(train_generator, validation_data=val_generator, epochs=3, callbacks=[early_stop])

# -----------------------------
# ✅ Plot Training History
# -----------------------------
def plot_history(histories, titles):
    for i, hist in enumerate(histories):
        # Accuracy
        plt.figure()
        plt.plot(hist.history['accuracy'])
        plt.plot(hist.history['val_accuracy'])
        plt.title(f'Accuracy - {titles[i]}')
        plt.legend(['Train', 'Val'])
        plt.show()

        # Loss
        plt.figure()
        plt.plot(hist.history['loss'])
        plt.plot(hist.history['val_loss'])
        plt.title(f'Loss - {titles[i]}')
        plt.legend(['Train', 'Val'])
        plt.show()

plot_history([history, history_fine], ["Feature Extraction", "Fine-Tuning"])

# -----------------------------
# ✅ Evaluation on Test Data
# -----------------------------
test_loss, test_acc = model.evaluate(test_generator)
print(f"\n✅ Test Accuracy: {test_acc:.4f}")

y_pred = (model.predict(test_generator) > 0.5).astype("int32")
y_true = test_generator.classes

classes = ["Uninfected","Parasitized"]

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title("Confusion Matrix")
plt.show()

print("\n🔍 Classification Report:\n")
print(classification_report(y_true, y_pred, target_names=classes))
```
